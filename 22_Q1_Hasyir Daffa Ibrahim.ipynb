```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Heart Attack Dataset

### a. Informasi Dataset

Link Dataset: https://www.kaggle.com/datasets/pritsheta/heart-attack/code

```python
df_heart = pd.read_csv('Heart Attack Data Set.csv')
```

```python
df_heart.head()
```

```python
df_heart.info()
```

```python
df_heart.describe()
```

### b. Visualisasi Dataset

```python
#Visualisasi distribusi umur pasien
plt.figure(figsize=(10,5))
plt.hist(df_heart['age'])
plt.title("Data Umur Pasien")
plt.xlabel("Age")
plt.ylabel("Jumlah")
plt.grid(True)
plt.show()
```

```python
#Visualisasi Rasio Pasien Terkena Serangan Jantung
#1 indikasi positif terkena
#0 indikasi negatif terkena
textprops = {"fontsize":22}
df_heart['target'].value_counts().plot(kind='pie',autopct="%.2f%%",textprops=textprops,radius = 1.2)
plt.ylabel('Target',fontsize=25)
```

```python

#Visualisasi Persebaran Data Numerik
df_heart.plot(kind='scatter',x="age", y="trestbps",color='#AF7AC5')
df_heart.plot(kind='scatter',x="age", y="chol",color='#F8C471')
df_heart.plot(kind='scatter',x="age", y="thalach",color='#82E0AA')
df_heart.plot(kind='scatter',x="age", y="oldpeak",color='#AAB7B8')
```

### c. Penjelasan

Terdapat 3 Model Visualisasi Data diatas yakni terdapat histogram, Pie chart dan Scatter plot. Dalam Histogram menggambarkan persebaran tentang data umur Pasien, dapat dilihat bahwa mayoritas pasien yang terdata berada pada umur kurang lebih 60 tahunan. Selanjutnya terdapat Pie Chart, disini menggambarkan ratio pasien yang terdeteksi Serangan jantung dan yang tidak, Hal ini ditandakan dengan "1" jika menderita dan "0" jika tidak. Dapat dilihat pada gambar Pie Chart tersebut bahwa rasionya relatif mendekati 50:50. Terkahir terdapat scatterplot, Scatterplot ini dilakukan terhadap data - data numerikal dalam dataset terhadap umur. Disini ingin melihat tentang korelasi Umur terhadap Restbps (tekanan darah), chol (serum cholesterol), thalac, dan oldpeak. Meskipun tidak dapat terlihat dengan jelas namun atribut - atribut ini tidak terlalu berkaitan dengan umur seorang pasien. Maka tujuan dari visualisasi ini dapat dijabarkan sebagai berikut:
1. Explanatory; Visualisasi ini menjelaskan penyebaran data umur seorang pasien yang melapor dan terdiagnosa oleh serangan jantung
2. Exploratory; Visualisasi ini menjelaskan bahwa serangan jantung dapat terjadi pada rentang umur yang cukup jauh dari yang masih muda sampai sudah tua. Serta, pengaruh dari atribut pada dataset terhadap diagnosa seorang pasien
3. Confirmatory; Hal yang menjadi konfirmasi dari beberapa visualisasi ini, bahwa umur seorang pasien tidak/hampir tidak berpengaruh terhadap serangan jantung seseorang. Melainkan faktor lain seperti tekanan darah, cholesterol, dan lainya.

## 2. India Food Prices Dataset

### a. Informasi Dataset

```python
df_income = pd.read_csv('income.csv')
df_income.head()

df_india = pd.read_csv('india_food_price.csv')
df_india.head()
```

```python
df_india.info()
```

```python
df_india['date'] = pd.to_datetime(df_india['date'])
date_min = df_india['date'].min()
date_max = df_india['date'].max()
```

### b. Visualisasi Standar + Desain Alternatif

```python
#Visualisasi Perubahan Harga Bawang di India
df_india.plot(x="date", y="Onions", kind="line", color="blue",ylabel="Count")

#Alternatif bar chart
df_india.plot(x="date", y="Onions", kind="bar", color="red",ylabel="Count")

#Alternatif area chart
df_india.plot(x="date", y="Onions", kind="area", color="green",ylabel="Count")

#Alternatif scatter plot
df_india.plot(x="date", y="Onions", kind="scatter", color="purple")
```

```python
#Visualisasi Perubahan Harga Nasi di India
df_india.plot(x="date", y="Rice", kind="line", color="blue",ylabel="Count")

#Alternatif bar chart
df_india.plot(x="date", y="Rice", kind="bar", color="red",ylabel="Count")

#Alternatif area chart
df_india.plot(x="date", y="Rice", kind="area", color="green",ylabel="Count")

#Alternatif scatter plot
df_india.plot(x="date", y="Rice", kind="scatter", color="purple")
```

```python
#Visualisasi Perubahan Harga Kentang di India
df_india.plot(x="date", y="Potatoes", kind="line", color="blue",ylabel="Count")

#Alternatif bar chart
df_india.plot(x="date", y="Potatoes", kind="bar", color="red",ylabel="Count")

#Alternatif area chart
df_india.plot(x="date", y="Potatoes", kind="area", color="green",ylabel="Count")

#Alternatif scatter plot
df_india.plot(x="date", y="Potatoes", kind="scatter", color="purple")
```

### Penjelasan

Pada bagian ini, Visualisasi utama yang digunakan adalah Line Chart. Karena data yang ingin di visualisasikan merupakan data Perubahan Harga Bawang (Onions), Nasi (Rice), dan Kentang (Potatoes) di India dari tahun ke tahun. Menurut Saya penggunaan Line Chart disini paling cocok karena Line chart sendiri secara dasar menggambarkan hubungan antara waktu dan kuantitas.
Kemudian disini terdapat alternatif visualisasi line chart yang lain berupa bar chart, area chart dan scatter plot

Kelebihan dan Kekurangan dalam Visualisasi Alternatif

**1. Bar Chart**
 
 Untuk visualisasi ini, penggunaan bar chart kurang baik digunakan karena untuk tahun ke tahun perubahannya walaupun mudah untuk dilihat melalui bar nya namun sulit untuk dilihat melalui tahunnya. Hal ini bisa di atasi tentunya jika dilakukan transformasi terhadap interval tahunnya. Namun hal tersebut menambahkan 1 langkah lagi dalam melakukan visualisasinya. Kelebihan yang bisa saya tangkap adalah dimana barchart dengan sangat detail menggambarkan suatu informasi. Detail disini bersifat relatif karena sesuatu hal yang terlalu detail bisa berakibatkan data yang terlalu sulit untuk dibaca.

 **2. Area Chart**

 Untuk Visualisasi ini, area chart sendiri kelebihannya adalah  cukup baik dalam merepresentasikan visual. Namun untuk kegunaan area chart sendiri tidak muncul kegunaannya pada data yang memiliki time series seperti perubahan harga bahan makanan di India ini. Karena biasanya area chart digunakan untuk menggambarkan perbandingan kuantitas antar atribut. Karena disini tujuan visualisasinya hanya untuk per-atribut , maka kurang cocok penggunaan area chart disini. 

 **3. Scatterplot**

 Untuk Visualisasi ini, Scatter plot merupakan pilihan yang paling kurang cocok untuk digunakan. Karena sifat Scatterplot adalah untuk melihat persebaran data, bukan untuk memvisualisasikan suatu hal yang berhubungan dengan time series. Seperti yang dapat dilihat pada scatterplot onions dan potatoes, agak sulit untuk dilihat kenaikan/penurunan harganya, justru yang terlihat adalah sedikit tersebar datanya. Hal yang bisa saya katakan menjadi kelebihan dari penggunaan scatterplot terhadap kasus ini adalah lebih mudah untuk menentukan harga tengah dari semua kenaikan dan penurunan tersebut


Kesimpulannya adalah dimana pada kasus ini, penggunaan terbaik adalah menggunakan Line Chart. Karena tujuan utama dalam visualisasi pada kasus ini adalah untuk menemukan kenaikan dan penurunan harga dari tahun ke tahun terhadap bahan makanan di India. Dengan menggunakan linechart, dapat kita lihat untuk 3 bahan makanan tersebut pada tahun 2019 sempat melonjak harganya untuk ketiganya, kemudian juga pada tahun 2014 - 2016 harganya paling sering berfluktuatif untuk kentang(Potatoes), dan bawang(Onions)

### c. Visualisasi &gt;2 Atribut

```python
#Visualisasi Harga yang melonjak
ax = df_india.plot(x="date", y="Sugar", kind="line",marker='.', color="#85929E",figsize=(15,10))
df_india.plot(x="date", y="Wheat", kind="line",marker='.',  ax=ax,color="#C39BD3")
df_india.plot(x="date", y="Chickpeas", kind="line",marker='.',  ax=ax,color="#5499C7")
df_india.plot(x="date", y="Rice", kind="line",marker='.', ax=ax,color="#AF7AC5")
plt.title("Perbandingan Perubahan Harga Makanan yang melonjak")
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid()
plt.show()
```

### Penjelasan

Pada bagian ini saya melakukan visualisasi terhadap harga Sugar, Wheat, Rice dan Chickpeas. Disini saya lakukan terhadap 4 atribut tersebut karena grafiknya yang bersifat naik melonjak pada kisaran tahun 2019. Dari visualisasi ini, kita dapat melihat bahwa untuk harga bahan makanan pokok yang bersifat karbohidrat ini mengalamni kelonjakan harga pada tahun 2019. Hipotesis saya terhadap hal ini adalah mungkin terjadi karena adanya covid - 19. Maka kita dapat mengetahui untuk bahan pokok saat terjadi covid - 19 harganya sempat menaik dengan drastis. Khususnya di India harga yang paling menaik adalah Wheat(gandum) dan Rice (nasi)

```python
#Visualisasi Harga yang relatif

ax = df_india.plot(x="date", y="Oil (mustard)", kind="line",marker='.',color="#82E0AA",figsize=(15,10))
df_india.plot(x="date", y="Potatoes", kind="line",marker='.', ax=ax,color="#C0392B")
df_india.plot(x="date", y="Onions", kind="line",marker='.', ax=ax,color="#AAB7B8")
plt.title("Perbandingan Perubahan Harga Makanan yang relatif sama")
```

### Penjelasan

Pada bagian ini Saya melakukan visualisasi terhadap harga Oil, Potatoes dan Onions. Disini saya lakukan terhadap 3 atribut tersebut karena grafiknya bersifat relatif tidak sangat berfluktuatif. Secara tampak awal, untuk harga Oil dapat dilihat menaik grafiknya, namun jika dibandingkan kuantitasnya pada grafik ini di sumbu y dengan grafik sebelumnya. Kenaikan ini bisa dikatakan tidak signifikan jika menggunakan parameter kenaikan 4 bahan pokok sebelumnya yang sudah dibahas. Tentu saja dibandingkan Potatoes dan Onions, harga Oil lebih mahal dari keduanya. Dari sini kita dapat dilihat pada tahun 2019, untuk harga 3 atribut ini tidak melonjak naik, beda dengan 4 atribut sebelumnya yang di hipotesiskan melonjak harganya pada tahun 2019 karena adanya covid - 19. Maka dengan adanya visualisasi ini, untuk membeli bahan makanan Oil, Potatoes dan Onions di India saat yang lainnya sedang naik harganya, ketiga ini tidak naik secara signifikan.  

### d. Faceting

```python
import seaborn as sns
g = sns.PairGrid(df_india)
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.histplot)
```

### Penjelasan

Pada Bagian ini dilakukan faceting. Faceting ini dilakukan karena biasanya karena semakin banyak adanya atribut maka semakin sulit untuk kita memahami suatu informasi. Maka dari itu faceting memudahkan kita untuk melakukan visualisasi terhadap masing - masing atribut sesuai keinginan kita pada sebuah dataset. 

Dapat dilihat pada hasil visualisasi diatas bahwa faceting dilakukan terhadap 2 atribut dengan model grafik scatterplot terhadap atribut yang berbeda. Disini kita dapat melihat perbedaan persebaran data antara tiap kedua atribut tersebut. Sayangnya, karena dataframe yang digunakan bukan merupakan data klasifikasi. Maka faceting disini hanya menggambarkan suatu persebaran. tidak terlalu menggambarkna suatu korelasi antar atribut. Karena pada konteksnya sendiri di dataframe ini adalah perubahan harga bahan makanan di India dari tahun ke tahun. Maka dataset yang tersedia pun juga tentang perubahan harga tersebut, jadi lebih kearah penggunaannya untuk melakukan prediksi ke masa depan, bukan untuk melakukan klasifikasi.
